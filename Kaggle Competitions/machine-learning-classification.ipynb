```python
import numpy as np
import pandas as pd
#Import matplotlib and seaborn for visualizations
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../input/machine-learning-classification/train.csv")
test = pd.read_csv("../input/machine-learning-classification/test.csv")
```

```python
df.columns = ['id', 'IP Mean', 'IP Sd', 'IP Kurtosis', 'IP Skewness','DM-SNR Mean', 'DM-SNR Sd', 'DM-SNR Kurtosis', 'DM-SNR Skewness', 'target_class']
test.columns = ['id', 'IP Mean', 'IP Sd', 'IP Kurtosis', 'IP Skewness','DM-SNR Mean', 'DM-SNR Sd', 'DM-SNR Kurtosis', 'DM-SNR Skewness']
```

```python
f,ax=plt.subplots(figsize=(15,15))
sns.heatmap(df.corr(),annot=True,linecolor="blue",fmt=".2f",ax=ax)
plt.show()
```

```python
X = df.drop(['target_class','id'], axis=1)
y = df['target_class']
```

```python
test = test.drop(['id'], axis=1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 1)
```

```python
X_train.shape, X_test.shape
```

```python
cols = X_train.columns
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)

X_test = scaler.transform(X_test)
```

```python
test = scaler.fit_transform(test)
```

```python
X_train = pd.DataFrame(X_train, columns=[cols])
X_test = pd.DataFrame(X_test, columns=[cols])
```

```python
from sklearn.ensemble import RandomForestClassifier

RFC =RandomForestClassifier(n_jobs=3,n_estimators=50,class_weight="balanced")
RFC.fit(X_train,y_train)
pred=RFC.predict(X_test)
#print(f'the Accuracy of the model is {round(accuracy_score(yval,pred)*100,4)}%')
```

```python
sub_preds =RFC.predict(test)
```
