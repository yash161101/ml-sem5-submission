## Row based filtering

* .loc - locate by name
* .iloc - locates by numerical index

```python
movie_df.loc['Sing']
```

```python
movie_df.iloc[3]
```

```python
sub_df = movie_df.loc['Prometheus': 'Sing']
sub_df
```

```python
sub_df = movie_df.iloc[1:4]
sub_df
```

```python
movie_df[(movie_df['revenue'] > 100) & (movie_df['metascore'] > 35)]
```

```python
directors_list = ['David Slade', 'James Gunn', 'Bryan Singer']
movie_df['director'].isin(directors_list) #get boolean series
```

```python
movie_df[movie_df['director'].isin(directors_list)] #get df rows that satisfy condition
```

```python
def rating_function(x):
    if x>=8.0:
        return 'good'
    else:
        return 'bad'
```

```python
movie_df['rating_category'] = movie_df['rating'].apply(rating_function)
movie_df.head()
```

```python
movie_df['rating_category'] = movie_df['rating'].apply(lambda x: 'good' if x > 8.0 else 'bad')
movie_df.head()
```
